```python
# 필요한 라이브러리 불러오기
import requests
from bs4 import BeautifulSoup
import re
import time
```

```python
lst = []


base_url = "https://www.fss.or.kr/fss/bbs/B0000206/list.do?menuNo=200690&bbsId=&cl1Cd=&pageIndex={}&sdate=&edate=&searchCnd=1&searchWrd="

for n in range(19):
  url = base_url.format(n+1)
  r = requests.get(url)
  soup = BeautifulSoup(r.text, "html.parser")
  examples = soup.select('tr > td.title > a')

  for example in examples:
    lst.append(example.get_text().strip())
```

```python
import pandas as pd
```

```python
ex = pd.DataFrame(lst, columns = ['대출사기형 사례'])
```
