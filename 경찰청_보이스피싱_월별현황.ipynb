- 2018~2022년 전체 월별 추이

```python
df_2018 = df[df['년']==2018]
df_2019 = df[df['년']==2019]
df_2020 = df[df['년']==2020]
df_2021 = df[df['년']==2021]
df_2022 = df[df['년']==2022]
```

```python
plt.figure(figsize=(10, 4))

sns.lineplot(data=df_2018, x='월',y='전화금융사기 발생건수', label='2018')
sns.lineplot(data=df_2019, x='월',y='전화금융사기 발생건수', label='2019')
sns.lineplot(data=df_2020, x='월',y='전화금융사기 발생건수', label='2020')
sns.lineplot(data=df_2021, x='월',y='전화금융사기 발생건수', label='2021')
sns.lineplot(data=df_2022, x='월',y='전화금융사기 발생건수', label='2022')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('<2018~2022년 전체 월별 추이>')
```

```python
# 주황색 그래프가 2019년 그래프 => 위쪽에 있는 것으로 보아 2018년부터 2022년까지 중 가장
# 전화금융사기 많았다는 것을 예측해볼 수 있고
# 이를 정확히 검증하기 위해 아래의 분석음 했습니다.
```

- 연도별 총 전화금융사기 발생 건수 변화

```python
# 각 연도별 총 전화금융사기 발생건수를 더했습니다.

df_2018_total = df_2018['전화금융사기 발생건수'].sum()
df_2019_total = df_2019['전화금융사기 발생건수'].sum()
df_2020_total = df_2020['전화금융사기 발생건수'].sum()
df_2021_total = df_2021['전화금융사기 발생건수'].sum()
# df_2022_total = df_2022['전화금융사기 발생건수'].sum()  -> 11월 12월 데이터가 없어서 제외함
```

```python
df_total = pd.DataFrame({2018:[df_2018_total], 2019:[df_2019_total], 2020:[df_2020_total],2021:[df_2021_total]})
df_total = df_total.transpose()
df_total.columns=['전화금융사기 발생건수']
df_total
```
